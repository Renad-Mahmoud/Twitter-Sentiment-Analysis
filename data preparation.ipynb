```python
import pandas as pd
d1 = pd.read_csv('/content/test1.csv')
d2 = pd.read_csv('/content/test2.csv')
d3 = pd.read_csv('/content/test3.csv')
d4 = pd.read_csv('/content/original.csv')
```

```python
# Merge all into one DataFrame
merged_df = pd.concat([d1, d2, d3, d4], ignore_index=True)
merged_df
```

```python
merged_df['Sentiment'].value_counts()
```

```python
# remove very short reviews (like 1-2 words)
# Function to count words
def word_count(text):
    return len(str(text).split())

# Filter rows where text has 3 or more words
df = merged_df[merged_df['Text'].apply(word_count) >= 3].reset_index(drop=True)
```

```python
df.shape
```

```python
df['Sentiment'].value_counts()
```

```python
# Downsample each class to 33,000 rows
balanced_df = (
    df.groupby('Sentiment', group_keys=False)
      .apply(lambda x: x.sample(n=33000, random_state=42))
      .reset_index(drop=True)
)
```

```python
balanced_df['Sentiment_Label'].value_counts()
```

```python
# shuffle and reset index
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)
balanced_df
```

```python
balanced_df.isna().sum()
```

```python
balanced_df.duplicated().sum()
```

```python
# Check text quality
balanced_df['Text'].apply(lambda x: len(str(x).split())).describe()
```
